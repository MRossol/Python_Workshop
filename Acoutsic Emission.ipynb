# Analyze AE Events

```python
AE_data = AE.AE_wavelets("Data/C-Glass_1perc_AE.dat", 41)
AE_data.get_SS("Data/C-Glass_1perc.csv")

SS_data = DIC.shift_SS(pd.read_csv("Data/C-Glass_1perc.csv", skiprows=8).values[:,[3,4]] * [100, 1])
```

```python
plt.line_plot(AE_data.get_plot_data('strains', 'amplitudes'),
              xlim=(0,1.05), ylim=(0,100),
              linestyles=('-',),
             xlabel='Strain (%)', ylabel=('AE Amplitude (dB)'))
```

```python
plt.dual_plot(SS_data, AE_data.get_plot_data('strains', 'index'),
             xlim=(0,1.05), ylim=((0,250),(0,250)),
              xlabel='Strain (%)', ylabel=('Stress (MPa)', 'Acoustic Events'),
             colors=plt.get_COLORS(('blue', 'red', 'black')),
              axis_colors=plt.get_COLORS(('blue', 'red')),
            linestyles=(('-', '--'), ('',)),
             markers=(('',), ('.',)),
             )
```
